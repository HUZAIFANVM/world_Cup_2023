# Data analysis (EDA)

#### most runs by a bowler in a single match

```python
#most runs conceded bowler in a single match 
df1=df.loc[df['runs'].idxmax(),['player','team','runs','run_rate','opponent']]
df1=df1.to_frame(name='Player_Info')
df1.reset_index(inplace=True)
df1
```

#### Highest wickets taken by bowler in a match

```python
#highest wickets taken by bowler in a match
df2=df.loc[df['wickets'].idxmax(),['player','team','wickets','opponent']]
df2=df2.to_frame(name='Player_Info').reset_index()
df2
```

#### highest number of maiden overs bowled in a match 

```python
# highest number of maiden overs bowled by in a match
df3=df.loc[df['maidens'].idxmax(),['player','team','maidens','opponent']]
df3=df3.to_frame(name='Player_Info')
df3
```

#### highest number of 6s conceded by a bowler in a single match

```python
#highest number of 6 in a match
df4=df.loc[df['6s'].idxmax(),['player','team','6s','opponent']]
df4=df4.to_frame(name='Player_Info')
df4
```

#### highest number of 4s conceded by a bowler in a match

```python
#highest number of 4s in a match 
df5=df.loc[df['4s'].idxmax(),['player','team','4s','opponent']]
df5=df5.to_frame(name='Player_Info')
df5
```

```python
df.columns
```

```python
df6=df.groupby('player').agg({'overs':'sum','maidens':'sum','runs':'sum','wickets':'sum','run_rate':'mean','6s':'sum','4s':'sum','wd':'sum','nb':'sum'})
df6=df6.sort_values(by='overs', ascending=False)
df6=df6.head(30)
df6=df6.reset_index()
df6
                          
```

### TOTAL OVERS AND MAIDENS BOWLED BY TOP 30 BOWLERS

```python
sns.set(style='white')
plt.figure(figsize=(15,6))
plt.title("Total Overs Bowled + Maidens By Top 30 Bowlers",fontsize=15)
plt.plot(df6['player'],df6['maidens'],marker='o',linestyle='--',color='red')
ax=sns.barplot(x='player',y='overs',data=df6)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45,ha='right')
plt.legend(['maidens','overs'])
plt.tight_layout()
plt.show()
```

### TOTAL RUNS GIVEN BY TOP 30 BOWLERS 

```python

plt.figure(figsize=(15,6))
ax1=sns.barplot(x='player',y='runs',color='green',data=df6)
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.xticks(rotation=70,ha='right')
plt.xlabel('Top 30 Players')
plt.ylabel('Runs')
plt.title('Total Runs Given By Top 30 Bowlers',fontsize=15)
plt.tight_layout()
plt.show()
```

### RUN RATE OF TOP 30 BOWLERS

```python
plt.figure(figsize=(15,6))
ax2=sns.barplot(x='player',y='run_rate',color='orange',data=df6)
for bars in ax2.containers:
    ax2.bar_label(bars,fmt='%.2f')
plt.title('Run_Rate of Top 30 bowlers',fontsize=15)
plt.xlabel('Players')
plt.ylabel('Run Rate')
plt.xticks(rotation=65,ha='right')
plt.tight_layout()
plt.show()
```

```python
df7=df6.sort_values(by='wickets', ascending=False)
```

### TOTAL NO OF OVERS AND WICKETS TAKEN BY THEM

```python
plt.figure(figsize=(15,6))
plt.plot(df6['player'],df6['wickets'],marker='o',linestyle='--',color='red')
ax3=sns.barplot(x='player',y='overs',data=df6,color='pink')
for bars in ax3 .containers:
    ax3.bar_label(bars)
plt.title('Total No Of Overs bowled + Wickets taken by top 30 bowlers',fontsize=15)
plt.xlabel('players')
plt.ylabel('count (wickets+overs)')
plt.xticks(rotation=65,ha='right')
plt.legend(['wickets','overs'])
plt.tight_layout()
plt.show()
```

### NUMBER OF SIXES CONCEIVED BY BOWLERS

```python
sns.set(style='whitegrid')
plt.figure(figsize=(15,6))
plt.plot(df6['player'],df6['6s'],marker='o',linestyle='--')
plt.title('Numbers Of Sixes Hit To Every Bowler',fontsize=15)
plt.xlabel('Player')
plt.ylabel('Number OF Sixes')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```
